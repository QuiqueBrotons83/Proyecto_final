```python
# Vamos a cargar las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
df = pd.read_csv("../data/data_procesed/amazon_data_combined.csv")
df.head()
```

```python
# Revisar dimensiones del dataset
df_shape = df.shape

# Revisar tipos de datos y valores nulos
df_info = df.info()

# Contar valores nulos por columna
df_nulls = df.isnull().sum()

# Contar duplicados
df_duplicates = df.duplicated().sum()

# Estadísticas descriptivas para variables numéricas
df_describe = df.describe()

df_shape, df_info, df_nulls, df_duplicates, df_describe
```

```python
# Vanos a arreglar las columnas que tienen datos tipo object que quiero hacerlos numericos.
# Creamos una copia del dataset para no modificar el original
df_clean = df.copy()
# Mostrar algunos valores únicos para entender el formato original
print("Valores únicos de ejemplo en 'actual_price':")
print(df_clean['actual_price'].dropna().unique()[:10])
```

```python
# Convertimos a string por si hay números mezclados con texto
df_clean['actual_price'] = df_clean['actual_price'].astype(str)
# Eliminamos simbolos de moneda, comas y espacios
df_clean['actual_price'] = df_clean['actual_price'].str.replace(r'[^\d.]', '', regex=True)
# Reemplazos strings vacios con NaN
df_clean['actual_price'] = df_clean['actual_price'].replace('', np.nan)
# Convertimos a float
df_clean['actual_price'] = df_clean['actual_price'].astype(float)
# Verifica el resultado
print("\nResumen después de limpieza:")
print(df_clean['actual_price'].describe())
print("Valores nulos:", df_clean['actual_price'].isnull().sum())
```

```python
# Hago lo mismo con la columna discount_price
df_clean['discount_price'] = df_clean['discount_price'].astype(str)
# Eliminamos simbolos de moneda, comas y espacios
df_clean['discount_price'] = df_clean['discount_price'].str.replace(r'[^\d.]', '', regex=True)
# Reemplazos strings vacios con NaN
df_clean['discount_price'] = df_clean['discount_price'].replace('', np.nan)
# Convertimos a float
df_clean['discount_price'] = df_clean['discount_price'].astype(float)
# Verifica el resultado
print("\nResumen después de limpieza:")
print(df_clean['discount_price'].describe())
print("Valores nulos:", df_clean['discount_price'].isnull().sum())
```

```python
# Limpio las columnas 'actual_price' y 'discount_price' de una sola vez
for col in ['actual_price', 'discount_price']:
    df_clean[col] = (
        df_clean[col]
        .astype(str)
        .str.replace(r'[^\d.]', '', regex=True)
        .replace('', np.nan)
        .astype(float))
```

```python
# Elimino la columna UnaUnnamed: 0
df_clean.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
# Convierto la columna 'no_of_ratings a int:
# Paso 1: Reemplazar comas, convertir a numérico (maneja NaN automáticamente)
df_clean['no_of_ratings'] = (
    df_clean['no_of_ratings']
    .astype(str)
    .str.replace(',', '', regex=True)
)

# Paso 2: Convertir a float primero para permitir NaNs
df_clean['no_of_ratings'] = pd.to_numeric(df_clean['no_of_ratings'], errors='coerce')

# Paso 3: Rellenar nulos si es necesario (aquí con 0, puedes cambiarlo por mediana si prefieres)
df_clean['no_of_ratings'] = df_clean['no_of_ratings'].fillna(0).astype(int)

```

```python
# Analizando los datos veo que estaria bien crear la columna 'discount_percentage' para ver el descuento que tiene el producto.
df_clean['discount_percent'] = (
    (df_clean['actual_price'] - df_clean['discount_price']) / df_clean['actual_price']
).round(2) * 100
```
