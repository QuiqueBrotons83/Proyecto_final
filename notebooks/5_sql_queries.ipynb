```python
# Vamos a cargar las librerías necesarias
from sqlalchemy import create_engine
import pandas as pd
```

```python
# Parámetros de conexión
usuario = "postgres"
contrasena = "quique"  # ← cambia si usas otra
host = "localhost"
puerto = "5432"
base_de_datos = "PROYECTO_FINAL_AMAZON"

# Crear conexión
conexion = f"postgresql+psycopg2://{usuario}:{contrasena}@{host}:{puerto}/{base_de_datos}"
engine = create_engine(conexion)
print("✅ Conectado a PostgreSQL")
```

```python
# Hago la primera query de dto por categoria
query = '''
SELECT main_category, ROUND(AVG(discount_percent)::numeric, 2) AS avg_discount
FROM amazon_products
GROUP BY main_category
ORDER BY avg_discount DESC;
'''
df_descuento = pd.read_sql_query(query, engine)
df_descuento.head()
```

```python
# Valoracion media por categoria
query = '''
SELECT main_category, ROUND(AVG(ratings::numeric), 2) AS avg_rating
FROM amazon_products
WHERE ratings ~ '^[0-9]+(\.[0-9]+)?$'  -- Ensure ratings contain only numeric values
GROUP BY main_category
ORDER BY avg_rating DESC;
'''
df_rating = pd.read_sql_query(query, engine)
df_rating.head()
```

```python
# Top 10 productos con más valoraciones
query = '''
SELECT name, no_of_ratings
FROM amazon_products
ORDER BY no_of_ratings::int DESC
LIMIT 10;
'''
df_top = pd.read_sql_query(query, engine)
df_top
```

```python
# Precio medio por subcategoria
query = '''
SELECT sub_category, ROUND(AVG(actual_price)::numeric, 2) AS avg_price
FROM amazon_products
WHERE actual_price > 0
GROUP BY sub_category
ORDER BY avg_price DESC;
'''
df_precios = pd.read_sql_query(query, engine)
df_precios.head()
```
